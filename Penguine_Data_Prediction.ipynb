## Linear Regression Model (Without Scikit-learn)
Here, we are using a dataset of different flipper Lenght and their respective Body Mass of penguines, The objective of the model is to predict the Body mass of the penguine using a simple linear regression.

```python
import pandas as pd

# Importing the sample dataset
data = pd.read_csv("https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/main/datasets/penguins_regression.csv")
data
```

```python
#Flipper Length (mm) vs Body Mass (g)

from matplotlib import pyplot as plt
data.plot(kind='scatter', x='Flipper Length (mm)', y='Body Mass (g)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

### Now lets estimate the value of weight(slope) and intercept

```python
data.plot(kind='scatter', x='Flipper Length (mm)', y='Body Mass (g)', s=32, alpha=0.8, label='Observed Body Mass (g)')

# Slope and intercept for the line plot
slope =  60 # Example value, replace with your actual slope
intercept = -8000  # Example value, replace with your actual intercept

# Generate the line data
flipper_length = data['Flipper Length (mm)']
predicted_mass_line = slope * flipper_length + intercept

# Plot the line
plt.plot(flipper_length, predicted_mass_line, color='red', label='Predicted Mass (Line)')

# Plot customization
plt.title('Flipper Length vs Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.grid(True)
plt.legend()

# Hide top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

plt.show()
```

```python
#defining linear model
def linear_model(flipper_weight, flipper_lenght, flipper_intercept):
  body_mass = flipper_weight * flipper_lenght + flipper_intercept
  return body_mass
```

```python
import numpy as np
#creating linear line and plotting
flipper_weight = 60 #predicted data from plot
flipper_lenght = np.array(data["Flipper Length (mm)"])
flipper_intercept = -8000 #predicted data from plot

prediction_mass = linear_model(
    flipper_weight,
    flipper_lenght,
    flipper_intercept
)
prediction_series = pd.Series(prediction_mass)
# Add the Series as a new column to your DataFrame
data['Predicted Mass'] = prediction_series
data
```

```python
# Ploting Predicted Mass in the graph

# Scatter plot
data.plot(kind='scatter', x='Flipper Length (mm)', y='Body Mass (g)', s=32, alpha=0.8, label='Observed Body Mass (g)')

# Line plot
flipper_length = data['Flipper Length (mm)']
predicted_mass = data['Predicted Mass']
plt.plot(flipper_length, predicted_mass, marker='o', linestyle='-', color='red', label='Predicted Mass')

# Plot customization
plt.title('Flipper Length vs Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.grid(True)
plt.legend()

# Hide top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

plt.show()
```

### To see the data in more refined way we can use xlim and ylim

```python
# Ploting Predicted Mass in the graph

# Scatter plot
data.plot(kind='scatter', x='Flipper Length (mm)', y='Body Mass (g)', s=32, alpha=0.8, label='Observed Body Mass (g)', color="black")

# Line plot
flipper_length = data['Flipper Length (mm)']
predicted_mass = data['Predicted Mass']
plt.plot(flipper_length, predicted_mass, linestyle='-', color='red', label='Predicted Mass')

# Plot customization
plt.title('Flipper Length vs Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.grid(True)
plt.legend()
plt.xlim(100, None)
plt.ylim(-3000, None)

# Hide top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

plt.show()
```
